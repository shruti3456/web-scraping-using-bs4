```python
from bs4 import BeautifulSoup
import requests
```

```python
# creating the url variable for sl website

url ="https://www.salesken.ai/"
```

```python
# access the url with request object 

result = requests.get(url)

# scraping the web page content 

webpage = result.content

# create a soup object for parsing the webpage using html

sl_soup_obj = BeautifulSoup(webpage,"html.parser")

print(type(sl_soup_obj))
```

```python
# close the request connection

result.close()
```

```python
# find out all the links present in the webpage
list_of_links =[]
for links in sl_soup_obj.find_all('a', href= True):
    print(links["href"])
    list_of_links.append(links["href"])
```

```python
print(list_of_links)
```

```python
import pandas as pd
df = pd.DataFrame(list_of_links,columns=["Hyperlinks of Salesken Webpage"])
df
```
